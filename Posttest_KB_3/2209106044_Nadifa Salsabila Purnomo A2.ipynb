Import Package

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np

Deklarasi File CSV

```python
df= pd.read_csv('Aemf1.csv')
df
```

Grafik Batang

```python
df['Cleanliness Rating'].value_counts()
```

```python
sns.set_style('darkgrid')

dfroom= df['Cleanliness Rating'].value_counts()
plt.figure(figsize=(10,5))
plt.xlabel('Rating')
plt.ylabel('Banyak')
plt.title ('Perbandingan Rating', fontsize=20, fontweight=600)
plt.bar(dfroom.index, dfroom, color=['green', 'red'])

plt.show()
```

Diagram batang di atas menunjukkan bahwa 25 rb pengunjung merasa puas namun ternyata ada 6 pengunjung yang merasa tidak puas

Pie Chart

```python
plt.figure(figsize=(10,5))
data = df.City.value_counts()
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.title('Perbandingan Airbnb Tiap Kota')

plt.show()
```

Berdasarkan pie chart diatas menunjukkan bahwa total airbnb terbanyak berada di Roma dan yang paling sedikit berada di Amsterdam

Grafik Histogram

```python
plt.figure(figsize=(10,5))
plt.hist(df['Room Type'], range=(1, 3))
plt.xlabel('Tipe Kamar')
plt.ylabel('Banyak')
plt.title('Perbandingan Tipe Kamar')
plt.show()
```

Dari histogram di atas menunjukkan bahwa lebih banyak tipe kamar apartemen/home dengan kapasitas lebih besar hingga banyaknya mencapai 25 rb lebih sedangkan tipe kamar share/berbagi dengan kapaasitas kecil yang ketersediaannya bahkan tidak mencapai 2000

Scatter Plot

```python
plt.figure(figsize=(10,5))
sns.scatterplot(df,
             x = "Cleanliness Rating",
             y = "Guest Satisfaction",
             hue = 'City'
             )
plt.title('Perbandingan Rating dan Kepuasan Pelanggan')
plt.show()
```

Berdasarkan scatter plot di atas menunjukkan bahwa rating kebersihan dan kepuasan pelanggan yang paling terbaik/tinggi berada di Vienna dan yang terendah berada di Roma

HeatMap

```python
city_total_corr = df.loc[:, 'Cleanliness Rating':'Bedrooms'].corr()
city_total_corr
```

```python
sns.heatmap(city_total_corr, cmap='OrRd', annot=True, center=0)
plt.show()
```

Berdasarkan heatmap diatas menunjukkan bahwa hubungan antara guest satisfaction dan cleanliness rating memiliki korelasi yang kuat dengan bedrooms
