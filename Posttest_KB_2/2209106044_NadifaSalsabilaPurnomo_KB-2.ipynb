<h4>Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi ke dalam dictionary. Laluditampilkan menggunakan prettytable untuk mempermudah tampilan data tabular agar lebih terstruktur</h4>

```python
dataem ={}

dataem['Jumlah Record'] = em.shape[0]
dataem['Jumlah Attribute'] = em.shape[1]
dataem['Nama Attribute'] = em.columns.tolist()
dataem['Jumlah Attribute Dengan Tipe Angka'] = em.select_dtypes(['int64','float64']).shape[1]

dictionarykey = list(dataem.keys())

table = prtl.PrettyTable()
table.add_column(dictionarykey[0], [dataem['Jumlah Record']])
table.add_column(dictionarykey[1], [dataem['Jumlah Attribute']])
table.add_column(dictionarykey[2], ['\n'.join(dataem['Nama Attribute'])])
table.add_column(dictionarykey[3], [dataem['Jumlah Attribute Dengan Tipe Angka']])

print(table)
```
